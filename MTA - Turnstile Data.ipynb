# Interpolation

## Part #1 Merge Date and Time

* starting time should be midnight Saturday

```python
import pandas as pd
import numpy  as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
turnstile = pd.read_csv("turnstile_160402.txt")
```

```python
stations = turnstile[["STATION", "C/A", "UNIT", "SCP", "LINENAME"]].drop_duplicates()
stations.shape
```

```python
# look up station data by route
N = turnstile['LINENAME'].str.contains('N')
```

```python
X = turnstile[N]

X['DATETIME'] = pd.to_datetime(turnstile[N]["DATE"] + " " + turnstile[N]["TIME"])
```

```python
X[ X.STATION == '57 ST-7 AV'].set_index('DATETIME')
```

```python
week = pd.date_range('2016-03-26 00:00:00', periods=24*7+1, freq='H')
```

```python
X
```

```python
Y = X[ X.STATION == '57 ST-7 AV'][(X.UNIT=='R080')&(X['C/A'] == 'A010')&(X.SCP=='00-00-00')]
Y = Y.set_index('DATETIME').reindex(week).interpolate()
Y['EXITS'] = Y['EXITS                                                               ']
Y[['ENTRIES', 'EXITS']].astype(int)
```
